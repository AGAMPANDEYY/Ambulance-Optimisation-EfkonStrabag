```python
import pandas as pd
import numpy as np

df_s=pd.read_csv("/workspaces/Accident-Opt-Strabag/DataFiles/sources_coord.csv")
df_dp=pd.read_csv("/workspaces/Accident-Opt-Strabag/DataFiles/KMeans_clusters.csv")
```

```python
distance_matrix=pd.DataFrame()

for i,row2 in df_s.iterrows():
    source_chainage=row2['Chainage']
    chainage_distance=[]
    for _, row in df_dp.iterrows():
        accident_chainage=row['Accident_Chainage']
        clockwise_chainage_distance= 156.7- abs(source_chainage-accident_chainage)  #maybe the shorter route  156.7 is the end Chainage km 
        anticlockwise_chainage_distance= abs(source_chainage-accident_chainage)  # based on chainage 

        #selecting the smaller distance 
        if clockwise_chainage_distance<anticlockwise_chainage_distance:
            chainage_distance.append(clockwise_chainage_distance)
        else:
            chainage_distance.append(anticlockwise_chainage_distance)
        
    distance_matrix[f's{i}-dp']=chainage_distance
```
