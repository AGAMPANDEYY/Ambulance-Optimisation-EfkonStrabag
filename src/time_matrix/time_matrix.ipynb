# Calculating time matrix based on Distance Matrix

## OSRM Distance Matrix

```python
import pandas as pd
import numpy as np

df_s_dp=pd.read_csv("/workspaces/Accident-Opt-Strabag/DataFiles/osrm_distance_matrix.csv")

df_s_dp.columns
```

```python
velocity=110.0/(60) #km/min
df_s_dp
df_time=pd.DataFrame()

for col in df_s_dp.columns:
    df_time[col]=df_s_dp[col]/velocity
```

```python
df_time
```

```python
df_time.to_csv("osrm_time_matrix.csv",index=False)
```
